### df.loc and df.iloc are different ways of performing the same task.
lets review how they work 
on the titanic dataset.

```python
#import the required library
import pandas as pd
```

```python
#load the dataset
df = pd.read_csv("full.csv")
#lets view our dataset by checking our top 5 entries
df.head()
```

### Note that : DataFrame.iloc is a Purely integer-location based indexing for selection by position.
###    and       : DataFrame.loc is a purely label-based indexing for selection by label.

```python
#lets review the DataFrame.loc first: this uses the rows and columns as its arguments , respectively:

# Retrieving the row number 3 with all its columns:
df.loc[3,:]

#note here that the indexing starts with 0 , so dom't get confused over that
```

```python
#similarily if you want to retrieve all the rows for a selected columns,
df.loc[:,'Survived'].head()
```

```python
#Lets try to populate multiple rows by passing them in the argument.
df.loc[[35,45,55],:]
```

```python
#Lets try to populate multiple columns by passing them in the argument.
df.loc[:,['Survived','Name','Ticket']].tail() #last 5 entries
```

```python
#we can also retrieve the first 7 rows using the :
df.loc[0:6,:]
```

```python
#We can also provide the range in the columns using the : operator, lets see how ?
#lets say we want the first 5 columns passenger_id till sex. 

df.loc[:,'PassengerId':'Sex'].head(10)
```

```python
#lets print the data only for the  passengers who have surivived. Survived == 1

df.loc[df.Survived==1,['Name','Sex']].head() #passing additional conditions.
```

Lets review the functionality of the iloc method now:
     Psst !  The 'i' above stands for integer

```python
#lets try getting the first 5 columns and 3 rows:
df.iloc[0:4,0:5]
#note here that the iloc method excludes the last value and the loc method includes it.
```

```python
df.loc[0:4,'PassengerId':'Sex']
#Note the difference above as the value at inde number 4 is included here and excluded above.
```

```python
#lets print 10 rows for all the columns except the last 4:
df.iloc[0:10,:-4]
```

```python
#See how this differs from the code above and wonder why.
#Hint : Negative indexing starts from the back.
df.iloc[0:10,-4]
```
